## **Statistical Testing: ANOVA**

```python
# Our first query for conducting statistic analysis (ANOVA)
query1 = """
SELECT borough, family_involvement_pct_positive, avg_student_attendance
FROM dim_environment e
JOIN dim_location l
ON e.DBN = l.DBN
"""

df1 = pd.read_sql_query(query1, conn)
df1
```

```python
# checking for NAs
df1.isna().sum()
```

```python
# dropping NAs (1st step for getting borough averages)
df_noNA = df1.dropna()
print(f'# of rows: {len(df_noNA)}')
print('---------------')
print(df_noNA.isna().sum())
```

```python
# grouping by borough (minus NAs) to get borough average

# individual group bys (for accurate averages)

avg_list = ['family_involvement_pct_positive', 'avg_student_attendance']
df_grouped = {}

for x in avg_list:
    df_grouped[x] = df_noNA.groupby('borough')[x].mean(numeric_only=False)

df_grouped
```

```python
# let's create a df where all boroughs are their averages, and then inner merge the original df1 to fill in the nulls with the borough average
df_family_avg = df1.copy()
df_family_avg['family_involvement_pct_positive'] = df_family_avg.groupby('borough')['family_involvement_pct_positive'].transform(lambda x: x.fillna(x.mean()))
df_attendance_avg = df1.copy()
df_attendance_avg['avg_student_attendance'] = df_attendance_avg.groupby('borough')['avg_student_attendance'].transform(lambda x: x.fillna(x.mean()))
df_family_avg
```
